```python
df = train.copy(deep=True)
```

```python
def cat_transform(value):
  return [i for i in value.split()[0]][0]
df['target']=df['target'].apply(cat_transform)
df['target']
```

```python
X = df.drop(columns=['target'])
Y = df['target']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
param_grid = {
    'n_estimators':[x for x in range(100,300,100)],
    'max_depth':[None,5,10],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4]
}
```

```python
from sklearn.model_selection import GridSearchCV
gs = GridSearchCV(estimator=rfc,param_grid=param_grid,cv=2)
gs.fit(xtrain,ytrain)



print(f'Best Hyperparameters: {gs.best_params_ }')
best_dt = gs.best_estimator_
print(best_dt)
```

```python
xtest
```

```python
pred = best_dt.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(pred,ytest)
```

```python
report=classification_report(pred,ytest,output_dict=True)
```

```python
report_df = pd.DataFrame(report)
report_df.to_csv('classification_report.csv')
```

```python
report = classification_report(pred,ytest)
print(report)
```

```python
sns.heatmap(confusion_matrix(pred,ytest),annot=True)
```

```python
df_test=test.copy(deep=True)
```

```python
df_test=sc.transform(df_test)
```

```python
act_pred=[]
for i in df_test:
  act_pred.append(best_dt.predict(i.reshape(1,-1)))
act_pred
```
